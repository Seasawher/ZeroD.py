```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    """ステップ関数"""
    return np.array(x > 0, dtype=int)

x = np.arange(-5, 5, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.show()
```

```python
def sigmoid(x):
    """シグモイド関数"""
    return 1 / (1 + np.exp(-x))

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
def relu(x):
    """ReLU関数"""
    return np.maximum(0, x)

x = np.arange(-5, 5, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```
