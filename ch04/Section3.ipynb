```python
def numerical_diff(f, x):
    """数値微分の悪い実装例

    差分の値が小さすぎて数値誤差が大きくなってしまう
    """
    h = 1e-50
    return (f(x+h) - f(x)) / (h)
```

```python
# 正しく 1.0 が返ってきてくれる
numerical_diff(lambda x: x, 0)
```

```python
# 6 が返ってきて欲しいが 0.0 が返ってくる
numerical_diff(lambda x: x**2, 3)
```

```python
def numerical_diff(f, x):
    """数値微分の悪い実装例 その２"""
    h = 1e-4
    return (f(x+h) - f(x)) / (h)

# ちゃんと6が返ってくる
numerical_diff(lambda x: x**2, 3)
```

```python
def numerical_diff(f, x):
    """数値微分の実装例"""
    h = 1e-4
    return (f(x+h) - f(x-h)) / (2 * h)

# 誤差が小さくなった
numerical_diff(lambda x: x**2, 3)
```

```python
import numpy as np
import matplotlib.pylab as plt

def function_1(x):
    return 0.01*x**2 + 0.1*x

x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)
plt.show()
```

```python
numerical_diff(function_1, 5)
```

```python
numerical_diff(function_1, 10)
```

```python
def tangent_line(f, x):
    """関数fのxでの接線の式 fun x => y を返す。"""
    d = numerical_diff(f, x)
    print(f"fのx={x}での微分の値: {d}")
    y = f(x) - d*x
    return lambda t: d*t + y

tf = tangent_line(function_1, 5)
y2 = tf(x)

plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```
